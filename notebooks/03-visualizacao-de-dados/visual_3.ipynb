```python
from matplotlib import pyplot as plt 
import seaborn as sns
import pandas as pd 
```

```python
df = sns.load_dataset("tips")
df.head()

sns.set_style("whitegrid")#linhas verticais e horizontais 
plt.figure(figsize=(6,3))
ax = sns.boxplot(x="day",y="total_bill",data=df,color='cornflowerblue')

plt.xlabel('Dia da semana')#plotando label eixo_x
plt.ylabel('Conta(USD)') #plotando label eixo_y
plt.title('Distribuição de gorjetas ao longo da semana')#plotando o titulo do grafico
plt.show() #plotagem do grafico completo
```
