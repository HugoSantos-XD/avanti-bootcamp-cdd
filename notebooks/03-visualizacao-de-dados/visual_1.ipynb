```python
from matplotlib import pyplot as plt 
import seaborn as sns
import pandas as pd 
```

```python
#LEITURA DOS DADOS

df = pd.read_csv('https://gist.githubusercontent.com/omadson/770934cb0df9e2d2c3e8db031f1139a5/raw/4d181efc035b327d9d66256915951384517c603e/arrecadacao.csv')
df.head()


df = df.assign(arrecadacao_milhao = df.arrecadacao / 1000000 ) #dividindo meus valores
df = df.sort_values(by = 'arrecadacao',ascending = False) #ordenando as barras 
sns.set_style("whitegrid") #linhas verticais

plt.figure(figsize =(5,2)) #plotando a dimensão do grafico

ax = sns.barplot(data=df, x='arrecadacao_milhao', y='titulo', color='tomato')

ax.bar_label( #valores nas barras
    ax.containers[0],
    fmt = '%.1f',
    label_type = 'center',
    color = 'white'
)

plt.xlabel("Arrecadação (Milhões de Dólares)") #label do eixo_X 
plt.ylabel("Filme") #label do eixo_y
plt.title("Arrecadação de filmes") #label do titulo do grafico

ax.set (frame_on=False)
ax.axvline(0,color="k",clip_on=False)
plt.show()#plotando meu grafico final



```
