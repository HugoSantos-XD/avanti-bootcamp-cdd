```python
from matplotlib import pyplot as plt 
import seaborn as sns
import pandas as pd 
```

```python
df = sns.load_dataset("penguins")
df.head()

sns.set_style("whitegrid")#linhas verticais e horizontais 
ax = sns.histplot(data=df, x='body_mass_g')
plt.ylabel('Quantidade')#plotando o eixo y 
plt.xlabel('Massa(g)') #plotando o eixo x 
plt.title('Distribuição de masssa dos pinguins')

ax.set(frame_on=False)
ax.axhline(0,color="k", clip_on= False)
plt.show()#plotando o grafico todo

```
